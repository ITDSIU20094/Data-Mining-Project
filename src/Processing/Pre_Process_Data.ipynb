## Data cleaning

### Subtask:
Clean the data by handling missing values, outliers, and duplicates.


**Reasoning**:
Impute missing values, handle outliers in 'price' and 'units_sold', remove duplicates, and validate the cleaning process.



```python
# Impute missing values
for col in df.select_dtypes(include=['number']):
    df[col] = df[col].fillna(df[col].median())
for col in df.select_dtypes(include=['object', 'category']):
    df[col] = df[col].fillna(df[col].mode()[0])

# Outlier handling for 'price' and 'units_sold'
for col in ['price', 'units_sold']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Data validation
print("\nMissing Values after cleaning:")
print(df.isnull().sum())

print("\nNumber of duplicates after cleaning:")
print(df.duplicated().sum())

# Outlier check after handling (example for 'price')
print("\nOutlier check for 'price' after handling:")
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(f"Number of outliers in 'price': {len(outliers)}")
```

## Self-training


### 💡 Why I Use Self-training and Choose the Threshold Units_Sold ≥ 5000

In this project, the goal is to build a classifier to predict whether a product is **hot-selling** based on features such as price, ratings, inventory, shipping options, and other product attributes.

#### ✅ Why choose Units_Sold ≥ 5000 as the labeling threshold?

I use the `units_sold` column to define a binary label (`hot_selling_label`):
- **Label = 1** if `units_sold ≥ 5000`, which indicates that the product has high sales volume.
- **Label = 0** otherwise.

I selected the threshold of 5000 because:
- It provides a reasonably large number of positive (label = 1) examples (~485), which is enough to train a supervised model effectively.
- The dataset remains relatively balanced between the two classes (Label 1 ≈ 36%, Label 0 ≈ 64%), which helps avoid bias in the classifier.
- This threshold draws a clearer line between truly popular products and others with lower sales, resulting in a more meaningful classification task.

#### 🔁 Why use Self-training?

Self-training is a semi-supervised learning approach where:
- A base classifier is first trained on a small set of labeled data (in this case, based on the `units_sold ≥ 5000` rule).
- The trained model is then used to predict labels for the unlabeled data, and the most confident predictions are added back into the training set to improve the model iteratively.

I chose Self-training because:
- While the initial labeled data (based on `units_sold`) is useful, there is still a large portion of the dataset that remains unlabeled or ambiguous (e.g., products with `units_sold = 1000` or `100`).
- Self-training helps utilize this unlabeled data more effectively, which can potentially **improve the model’s performance and generalization**.
- This method reflects a realistic scenario where manual labeling is expensive or impractical, and we want to leverage available data efficiently.

In summary, Self-training is a practical and effective method in this context, and the threshold of `units_sold ≥ 5000` provides a solid starting point for generating reliable initial labels.
